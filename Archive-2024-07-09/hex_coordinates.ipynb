### use this hex coordinates to compare neighboring ommatidia values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv file
hex_coord = pd.read_csv(
    "~/Documents/GitHub/CompoundRayTests/MotionDetector/nb_co.csv"
)  # unit hexagon coordinates
ind_nb = pd.read_csv(
    "~/Documents/GitHub/CompoundRayTests/MotionDetector/ind_nb.csv"
)  # index of neighbors
ind_xy = pd.read_csv(
    "~/Documents/GitHub/CompoundRayTests/MotionDetector/ind_xy.csv"
)  # hex coordinates of neighbors
```

```python
# plot xy coordinates
plt.figure(figsize=(10, 10))
plt.scatter(ind_xy["x"], ind_xy["y"], s=1)
# color points red if its x coordinate is divisible by 3
plt.scatter(
    ind_xy["x"][ind_xy["x"] % 3 == 0], ind_xy["y"][ind_xy["x"] % 3 == 0], c="r", s=3
)
# add coordinate text for every other point
for i in range(0, len(ind_xy)):
    if ind_xy["x"][i] % 3 == 0:
        plt.text(
            ind_xy["x"][i],
            ind_xy["y"][i],
            str([ind_xy["x"][i], ind_xy["y"][i]]),
            fontsize=8,
        )

plt.show()
```
